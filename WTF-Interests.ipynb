```python
from collections import Counter
import re

no_paren = re.compile('\([^)]*\)')

def clean(item):
    return no_paren.sub('', item).strip()

topics = Counter()
for line in people_topics.iloc[:, 1]:
    if pandas.isnull(line):
        continue
    line = line.lower().replace(',', '.')
    parts = (clean(s) for s in line.split('.') if s.strip())
    terms = []
    for part in parts:
        if part in related:
            terms.append(part+' -> '+related[part])
            terms.append(related[part])
        else:
            terms.append(part)
    topics.update(terms)

groups = []
current_group = []
current_count = 0
for topic, count in topics.most_common():
    if count != current_count:
        current_count = count
        groups.extend(sorted(current_group))
        current_group = []
    current_group.append((count, topic))
else:
    groups.extend(sorted(current_group))

for group in groups:
    print(*group, sep='\t')
```
